```python
pd.set_option('display.max_columns',None)
df = pd.read_csv("/kaggle/input/academic-scores/database.csv")
df
```

```python
sport = "Men's Basketball"
dfbb = df[df["SPORT_NAME"]==sport]
dfbb
```

```python
ID='SCHOOL_NAME'
y = dfbb['FOURYEAR_SCORE']
X_Num = dfbb[[ID,'FOURYEAR_ATHLETES','FOURYEAR_ELIGIBILITY','FOURYEAR_RETENTION']]
```

```python
SUB = pd.get_dummies(dfbb['NCAA_SUBDIVISION'],dtype='int')
SUB_1 = SUB[1]
SUB_1
```

```python
X = pd.concat([X_Num,SUB_1],axis=1)
X
```

```python
data = pd.concat([X,y],axis = 1)
data
```

```python
data.info()
```

```python
data.describe()
```

```python
xv = 'FOURYEAR_ELIGIBILITY'
px.scatter(data_frame = data, x=xv,y='FOURYEAR_SCORE',trendline='ols',hover_name=ID,hover_data=[xv,'FOURYEAR_SCORE'])
```

```python
px.scatter_3d(data_frame = data, x='FOURYEAR_RETENTION',y='FOURYEAR_ELIGIBILITY',z='FOURYEAR_SCORE',color = 1,hover_name=ID)
```

```python
# Add a constant to get an intercept
X1 = sm.add_constant(X.drop(ID,axis=1))
# Fit the regression line using ‘OLS’
mlr = sm.OLS(y, X1).fit()

print(mlr.summary())
```
